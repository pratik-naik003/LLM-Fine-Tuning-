Creating a custom dataset

```python
data={
    "text":[
        "I love machine learning",
        "HUgging face is amazing",
        "Transformers are powerful",
        "I enjoy learning AI",
        "Datasets are important",
        "Deep learning is facinating"
    ],
    "label":[1,1,1,1,0,1]
}
print(data)
```

convert the dataset into pandas dataframe

```python
import pandas as pd
df=pd.DataFrame(data)
print(df)
```

convert the padas dataset into huggingface dataset

```python
!pip install dataset
```

```python
from datasets import Dataset
hf_dataset=Dataset.from_pandas(df)
print(hf_dataset)
```

convert label into classlabel

```python
from datasets import ClassLabel
hf_dataset=hf_dataset.cast_column("label",ClassLabel(names=["negative","positive"]))
```

```python
print(hf_dataset)
```

Load Tokenizer

```python
from transformers import AutoTokenizer
```

```python
tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased")
```

Tokenize a sentences

```python
sentence="hello world"
tokens=tokenizer.tokenize(sentence)
print(tokens)
```

Concvert token into IDs

```python
encoded=tokenizer(sentence,return_tensors="pt")
print(encoded)
```

tokenizing entire dataset

```python
def tokenize(example):
  return tokenizer(
      example['text'],
      padding="max_length",
      truncation=True
  )
```

```python
encoded_dataset=hf_dataset.map(tokenize,batched=True)
```

```python
print(encoded_dataset)
```

Fast Tokenization

```python
fast_tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased",
                                             use_fast=True)
```

Compare fast VS Slow

```python
import time

text = ["Hugging Face is amazing"] * 10000

# fast tokenizer
start = time.time()

fast_tokenizer(text, padding=True, truncation=True)

print("Fast tokenizer time:", time.time() - start)
```

Slow tokenizer

```python
slow_tokenizer = AutoTokenizer.from_pretrained(

    "bert-base-uncased",

    use_fast=False
)

start = time.time()

slow_tokenizer(text, padding=True, truncation=True)

print("Slow tokenizer time:", time.time() - start)
```
